```python
import scipy.io
import matplotlib.pyplot as plt

# Load the .mat file
data = scipy.io.loadmat('brain_data.mat')

# Get the connectivity matrices
SCmatrices = data['SCmatrices']

# Plot the connectivity matrices
for i in range(len(SCmatrices)):
    plt.imshow(SCmatrices[i])
    plt.title('Connectivity Matrix {}'.format(i))
    plt.show()
```
